### Using a Grid Search

About the grid Search

```python
grid = GridSearchCV(
    LogisticRegression(),
    param_grid={},
    cv=CV,
    return_train_score=True,
    refit=True,
    n_jobs=N_JOBS,
    verbose=2,
)
```

Fit : 

```python
grid.fit(X_train, y_train)
```

Our results : 

```python
grid.cv_results_
```

In a dataframe : 

```python
pd.DataFrame(grid.cv_results_)
```

Lets's create a function : 

```python
def resultize(grid):

    res = grid.cv_results_
    res = pd.DataFrame(res)

    cols = [i for i in res.columns if "split" not in i]
    res = res.loc[:, cols]

    res = res.drop(columns=["mean_score_time", "std_score_time"])

    return res.round(2).sort_values("mean_test_score", ascending=False)
```

Resultize : 

```python
resultize(grid)
```

### Using a pipeline

Our first pipeline : 

```python
pipe = Pipeline(
    [
        ("imputer", KNNImputer()),
        ("scaler", StandardScaler()),
        ("estimator", LogisticRegression()),
    ]
)
```

```python
pipe
```

Using the pipeline : 

```python
grid = GridSearchCV(
    pipe,
    param_grid={},
    cv=CV,
    return_train_score=True,
    refit=True,
    n_jobs=N_JOBS,
    verbose=2,
)
```

Fit : 

```python
grid.fit(X_train, y_train)
```

Resultize : 

```python
resultize(grid)
```

### Using a Param Grid

Keep in mind : 

```python
pipe
```

```python
grid
```

Writing a beautiful param grid : 

```python
param_grid = {
    "imputer": [
        "passthrough",
        KNNImputer(n_neighbors=3),
        KNNImputer(n_neighbors=5),
        SimpleImputer(strategy="median"),
    ],
    "scaler": [
        # "passthrough",
        StandardScaler(),
        Normalizer(),
        QuantileTransformer(n_quantiles=10),
    ],
    "estimator": [
        DummyClassifier(),
        LogisticRegression(),
        KNeighborsClassifier(n_neighbors=5),
        RandomForestClassifier(),
    ],
}
```

```python
param_grid
```

Using the param grid : 

```python
grid = GridSearchCV(
    pipe,
    param_grid=param_grid,
    cv=CV,
    return_train_score=True,
    refit=True,
    n_jobs=N_JOBS,
    verbose=1,
)
```

Fit : 

```python
grid.fit(X_train, y_train)
```

Resultize ! 

```python
resultize(grid).head(10)
```
